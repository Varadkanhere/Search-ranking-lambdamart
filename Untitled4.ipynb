```python
import pandas as pd
import numpy as np

def load_mslr(filepath):
    data = []
    with open(filepath, 'r') as f:
        for line in f:
            parts = line.strip().split()
            label = int(parts[0])
            qid = int(parts[1].split(':')[1])
            features = [float(p.split(':')[1]) for p in parts[2:]]
            data.append([label, qid] + features)

    cols = ['label', 'qid'] + [f'f{i}' for i in range(1, 137)]
    return pd.DataFrame(data, columns=cols)

print("Loading test.txt...")
test_df = load_mslr('test.txt')

print("Loading vali.txt...")
val_df = load_mslr('vali.txt')

print("Loading train.txt...")
train_df = load_mslr('train.txt')

print("Done!")
print(f"Train shape: {train_df.shape}")
print(f"Val shape: {val_df.shape}")
print(f"Test shape: {test_df.shape}")
```

```python
def ndcg_at_k(y_true, y_score, k=10):
    order = np.argsort(y_score)[::-1]
    y_true_sorted = np.array(y_true)[order[:k]]

    gains = 2**y_true_sorted - 1
    discounts = np.log2(np.arange(2, len(gains)+2))
    dcg = np.sum(gains / discounts)

    ideal_sorted = np.sort(y_true)[::-1][:k]
    ideal_gains = 2**ideal_sorted - 1
    ideal_dcg = np.sum(ideal_gains / discounts[:len(ideal_gains)])

    return dcg / ideal_dcg if ideal_dcg > 0 else 0.0

def evaluate_ndcg(df, score_col, k=10):
    ndcg_scores = []
    for qid, group in df.groupby('qid'):
        if len(group) < 2:
            continue
        score = ndcg_at_k(group['label'].values, group[score_col].values, k)
        ndcg_scores.append(score)
    return np.mean(ndcg_scores)

# BM25 baseline
test_df['bm25_score'] = test_df['f1'] + test_df['f2']
bm25_ndcg = evaluate_ndcg(test_df, 'bm25_score', k=10)
print(f"BM25 Baseline NDCG@10: {bm25_ndcg:.4f}")
```

```python
!pip install lightgbm

import lightgbm as lgb

feature_cols = [f'f{i}' for i in range(1, 137)]

X_train = train_df[feature_cols].values
y_train = train_df['label'].values
q_train = train_df.groupby('qid').size().values

X_val = val_df[feature_cols].values
y_val = val_df['label'].values
q_val = val_df.groupby('qid').size().values

X_test = test_df[feature_cols].values
y_test = test_df['label'].values
q_test = test_df.groupby('qid').size().values

train_data = lgb.Dataset(X_train, label=y_train, group=q_train)
val_data   = lgb.Dataset(X_val,   label=y_val,   group=q_val)

params = {
    'objective': 'lambdarank',
    'metric': 'ndcg',
    'ndcg_eval_at': [10],
    'learning_rate': 0.05,
    'num_leaves': 63,
    'min_data_in_leaf': 50,
    'verbose': 25
}

print("Training LambdaMART... (takes 10-15 mins)")
model = lgb.train(
    params,
    train_data,
    num_boost_round=500,
    valid_sets=[val_data],
    callbacks=[lgb.early_stopping(30), lgb.log_evaluation(25)]
)

print("Training complete!")
```

```python
test_preds = model.predict(X_test)
test_df['lambdamart_score'] = test_preds

lambdamart_ndcg = evaluate_ndcg(test_df, 'lambdamart_score', k=10)

print("=" * 45)
print(f"  BM25 Baseline NDCG@10  : {bm25_ndcg:.4f}")
print(f"  LambdaMART NDCG@10     : {lambdamart_ndcg:.4f}")
improvement = ((lambdamart_ndcg - bm25_ndcg) / bm25_ndcg) * 100
print(f"  Improvement            : +{improvement:.1f}%")
print("=" * 45)
```

```python
import matplotlib.pyplot as plt

feature_names = {
    'f1': 'BM25 Title', 'f2': 'BM25 Body', 'f3': 'BM25 Anchor',
    'f4': 'TF Title', 'f5': 'TF Body', 'f6': 'IDF Title',
    'f7': 'PageRank', 'f8': 'URL Length', 'f9': 'Inlink Count',
    'f10': 'Outlink Count'
}

importance = model.feature_importance(importance_type='gain')
feat_imp_df = pd.DataFrame({
    'feature': feature_cols,
    'importance': importance
}).sort_values('importance', ascending=False).head(15)

feat_imp_df['name'] = feat_imp_df['feature'].map(
    lambda x: feature_names.get(x, x)
)

plt.figure(figsize=(10, 6))
plt.barh(feat_imp_df['name'][::-1], feat_imp_df['importance'][::-1], color='steelblue')
plt.xlabel('Feature Importance (Gain)')
plt.title('Top 15 Most Important Ranking Features\n(LambdaMART on MSLR-WEB10K)')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=150)
plt.show()
```

```python
def show_ranked_results(qid, model, df, feature_cols, top_k=10):
    query_docs = df[df['qid'] == qid].copy()

    query_docs['bm25'] = query_docs['f1'] + query_docs['f2']
    query_docs['ltr_score'] = model.predict(query_docs[feature_cols].values)

    print(f"\n{'='*55}")
    print(f"Query ID: {qid}  |  Total docs: {len(query_docs)}")
    print(f"{'='*55}")

    print(f"\n❌ BM25 Baseline (Top 10):")
    print(f"{'Rank':<6} {'Relevance':<20} {'Score':<12}")
    print("-" * 38)
    for i, (_, row) in enumerate(query_docs.nlargest(top_k, 'bm25').iterrows(), 1):
        stars = '⭐' * int(row['label']) if row['label'] > 0 else '💀 irrelevant'
        print(f"{i:<6} {stars:<20} {row['bm25']:.3f}")

    print(f"\n✅ LambdaMART Ranked (Top 10):")
    print(f"{'Rank':<6} {'Relevance':<20} {'Score':<12}")
    print("-" * 38)
    for i, (_, row) in enumerate(query_docs.nlargest(top_k, 'ltr_score').iterrows(), 1):
        stars = '⭐' * int(row['label']) if row['label'] > 0 else '💀 irrelevant'
        print(f"{i:<6} {stars:<20} {row['ltr_score']:.3f}")

sample_qid = test_df['qid'].iloc[100]
show_ranked_results(sample_qid, model, test_df, feature_cols)
```
